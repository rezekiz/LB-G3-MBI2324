### Alergia Alimentar: uma visão imunológica
A alergia alimentar é um problema crescente de saúde pública a nível mundial sendo a patofisiologia do seu desenvolvimento uma complexa interação de fatores ambientais e fatores genéticos. Após a exposição a uma proteína alimentar alergénica, o sistema imunitário desencadeia uma resposta através da produção de IgE antigénio específica que, numa reexposição, leva à ligação do alimento alergénico a anticorpos específicos e são libertados mediadores, como a histamina, levando a sintomalogia característica da alergia alimentar (Kanchan et al., 2021).

### Complexo HLA: análise e papel na respota imunitária adaptativa
O complexo antigénio leucocitário humano (HLA) tem sido uma das regiões genéticas mais frequentemente estudadas no que diz respeito a alergia alimentar e os resultados obtidos através da investigação do complexo HLA revelaram-se muito heterogéneos, principalmente devido à alta taxa polimórfica da região, ao alto linkage disequilibrium, associação não-aleatória de alelos em dois ou mais loci, e às diferentes classes de variantes (Megiorni & Pizzuti, 2012; Suaini et al., 2019). Este complexo é codificado pelo conjunto de genes HLA que se localiza no cromossoma 6 e cobre uma área de 7.6 Mb. O conjunto de genes HLA é composto por mais de 250 genes com funções relacionadas com resposta, regulação e vigilância do sistema imunitário sendo responsável por codificar famílias de proteínas da superfície celular que funcionam como determinantes-chave no reconhecimento de antigénios pelo sistema imunitário adaptativo. O complexo HLA encontra-se dividido em três classes de antigénios de acordo com a sua distribuição e função: classe I, classe II e classe III. Os antigénios da classe II incluem as moléculas HLA-DQA1, HLA-DQA2 e HLA-DQB1(Kanchan et al., 2021; Liu et al., 2021).

As moléculas da classe II desempenham um papel fundamental na resposta imunitária adaptativa, são glicoproteínas transmembranares que apresentam péptidos ao recetor de antigénio dos linfócitos T CD4+ e participam na seleção do reportório de células T, sendo cruciais no auto-reconhecimento por células imunitárias (Megiorni & Pizzuti, 2012; Miyadera & Tokunaga, 2015; Zajacova et al., 2015). 

###  A conexão do complexo HLA e a Doença Celíaca
Uma das doenças autoimunes mais prevalentes e em que o complexo HLA tem um papel preponderante é a doença celíaca (DC). Esta doença consiste numa perturbação multifatorial comum do intestino delgado causada por uma resposta imunitária inadequada ao glúten de trigo ingerido e a proteínas semelhantes do centeio e da cevada. Em indivíduos afetados, a ingestão de glúten leva a inflamação crónica da mucosa do intestino delgado, levando à manifestação de sintomas, como: diarreia, perda de peso e, até, anomalias dentárias (Tabela 1).  A prevalência da DC está estimada em 1 a cada 100 caucasianos, mas, em muitos casos, permanece não diagnosticada devido à ampla variedade de sintomas (Caillat-Zucman, 2009; Megiorni & Pizzuti, 2012).

### Alergia ao amendoim e a relação com o complexo HLA
Uma das associações existentes entre a alergia alimentar e as moléculas da classe II do complexo HLA passa pela alergia ao amendoim. O amendoim (Arachis hypogea) é um dos alimentos mais comuns a causar reações alérgicas agudas, tanto em crianças como em adultos e é a causa mais frequente de anafilaxia fatal e anafilaxia quase fatal induzida por alimentos (Sampson et al., 1992). Atualmente, pensa-se que esta alergia afeta cerca de 1 a 3% das crianças (Sicherer & Sampson, 2018), estando a sua incidência fortemente relacionada com fatores genéticos e ambientais (Abrams et al., 2020). Crianças com parentes próximos que possuam reação alérgica aos amendoins, possuem uma suscetibilidade maior para esta perturbação, bem como as crianças do sexo masculino comparativamente ao sexo feminino (Begin et al., 2016).

As reações alérgicas provocadas pelo amendoim podem ser induzidas por ingestão ou contato, desenvolvem-se numa fase inicial da vida e normalmente persistem por tempo indefinido (Keating et al., 1990). Um dos principais problemas para o doente, é que os amendoins são particularmente difíceis de evitar, pois encontram-se presentes numa grande variedade de alimentos, por vezes até disfarçados de outros frutos secos. Todos estes fatores fazem com que esta reação alérgica seja um fardo diário que afeta significativamente e negativamente a qualidade de vida das pessoas afetadas (Sicherer & Sampson, 2010).

De acordo com a revisão sistemática de Suaini et al. (2019), a relação entre as moléculas DQB1 e DRB1 e a alergia ao amendoim é altamente reprodutível (Suaini et al., 2019). Adicionalmente, recentemente foi descoberta a associação entre o alelo HLA-DQA1*01:02 e uma IgE (específica para o Ara h 2), desempenhando um papel crucial no desenvolvimento da alergia ao amendoim (Kanchan et al., 2022).
